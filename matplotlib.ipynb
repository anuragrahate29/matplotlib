```python
url = "https://raw.githubusercontent.com/anuragrahate29/matplotlib/refs/heads/main/fifa_data.csv"
fifa = pd.read_csv(url)
fifa.head()
```

```python
bins = [40,50,60,70,80,90,100]
#bins in histogram is used to collect all the datapoints (here Overall) that fall in the range
# given by bins
# which means that around 10000 players fall under the range of 60-70 Overall points
plt.hist(fifa.Overall,bins=bins,color='#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Overall Points Range')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
left
```

```python
(fifa['Preferred Foot']=='Left').sum()
# The above query only check the 'Preferred column' with the condition , counts the true(ie 1) and returns value
# while 'fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]' query checks every row of the dataset , returns the list of all columns having the count of
# not null values .
```

```python
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
```

```python
labels = ['Left','Right']
colors = ['#ddbbcc','#eeccdd']
plt.pie([left,right],labels = labels,colors = colors, autopct ='%.2f%%')

plt.show()
```

```python
fifa.Weight = fifa['Weight'].apply(lambda x: int(x.strip('lbs')) if isinstance(x, str) else x)
fifa.Weight
```

```python

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']
colors = ['#ddbbcc','#eeccdd','#aabbcc','#bbbbcc','#ccbbcc']
explode = [0.5,0.1,0,0,0.5]
plt.pie(weights,labels=labels,autopct='%.2f%%',pctdistance = 0.8 ,colors = colors,explode = explode)
plt.title('Weight Distribution of FIFA Players in lbs')
plt.show()
```
