```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#plt.plot(x,y, label='2x',color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')
plt.show
#Use shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y,'b.--',label='2x')
plt.title('My first plot')
plt.xlabel('x axis')
plt.ylabel('y axis')

#Line number two
x2 = np.arange(0,4.5,0.5)
#plt.plot(x2,x2**2,'r',label='x^2')
plt.plot(x2[:5],x2[:5]**2,'r',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'r--')
plt.legend()

plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,2,4,6,8,10])
plt.show()
#plt.savefig('mygraph.png',dpi=300) The dpi is used to increase/decrease the clarity and size of the graph
#dpi 300 is good when saving graph with high resolution

```

```python
labels = ['A','B','C']
values = [2,4,3]
bars = plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('|')
plt.yticks(np.arange(0,max(values)+1,1))
plt.show()
```
